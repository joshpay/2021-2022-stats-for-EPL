# HIGHEST GOALSCORERS 2021/2022 SEASON


```python
#The top 10 goalscorers and their goals scored
goalscorer_df = pl_df[['Player', 'Team','Pos','Ast','Gls','Age','MP','Min','90s','Starts','xG','xG.1']].sort_values('Gls', ascending= False).head(10)
goalscorer_df
```

```python
font =  {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
```

```python
plt.subplots(figsize =(20, 8))
plt.title('Goals and Assists by the Highest Goals scorers', fontdict = font)
plt.bar(goalscorer_df.Player, goalscorer_df.Gls, width=0.4, color='r', label='Goals' )
plt.bar(goalscorer_df.Player, goalscorer_df.Ast, width=0.4, color='b', label= 'Assists',bottom = goalscorer_df.Gls)
plt.ylabel('Goals and Assists',fontsize=20)
plt.xlabel('Players Name',fontsize=20);
plt.legend()
plt.show()
```

# Penalties Merchants










Players who have played the most penalties.

```python
PK_merchs_df = pl_df[['Player','Team', 'Pos', 'Gls','GPK','PK','PKatt']]
PK_merchs_df = PK_merchs_df.sort_values(['PK','PKatt'], ascending=False).head(10)
PK_merchs_df
```

```python
plt.subplots(figsize =(20, 8))
plt.title('Goals By Penalty Merchants', fontdict = font)
plt.bar(PK_merchs_df.Player, PK_merchs_df.GPK, width=0.4, color='skyblue', label='Goals without PK' )
plt.bar(PK_merchs_df.Player, PK_merchs_df.PK, width=0.4, color='r', label= 'PK Goals',bottom = PK_merchs_df.GPK)
plt.ylabel('Goals',fontsize=20)
plt.xlabel('Players Name',fontsize=20);
plt.legend()
plt.show()
```
